### Matrix Multiplication

Matrix multiplication is an operation that takes a pair of matrices, and produces another matrix. It's important to note that for matrix multiplication to be defined, the number of columns in the first matrix must be equal to the number of rows in the second matrix.

Suppose we have two matrices \( A \) and \( B \):

\[ A = \begin{bmatrix}
a_{11} & a_{12} & \ldots & a_{1k} \\
a_{21} & a_{22} & \ldots & a_{2k} \\
\vdots & \vdots & \ddots & \vdots \\
a_{m1} & a_{m2} & \ldots & a_{mk} \\
\end{bmatrix} \]

\[ B = \begin{bmatrix}
b_{11} & b_{12} & \ldots & b_{1n} \\
b_{21} & b_{22} & \ldots & b_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
b_{k1} & b_{k2} & \ldots & b_{kn} \\
\end{bmatrix} \]

The result of the multiplication, denoted \( C \), is a matrix of size \( m \times n \):

\[ C = A \cdot B = \begin{bmatrix}
c_{11} & c_{12} & \ldots & c_{1n} \\
c_{21} & c_{22} & \ldots & c_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
c_{m1} & c_{m2} & \ldots & c_{mn} \\
\end{bmatrix} \]

The elements \( c_{ij} \) in the resulting matrix are computed as follows:

\[ c_{ij} = a_{i1} \cdot b_{1j} + a_{i2} \cdot b_{2j} + \ldots + a_{ik} \cdot b_{kj} \]

__Hence for each row in matrix A we have k multiplications, we then have these multiplications on m columns. hence a total of m\*k multiplications for each row. We have n rows and hence total number of multiplications for matrix multiplication required are n\*k\*m.
similarly we have m*(k-1) addition operations for each row which implies total number of addition operation to be n\*(k-1)\* m__

Matrix multiplication is not commutative, meaning that \( A \cdot B \) is not necessarily equal to \( B \cdot A \). Also, the number of columns in matrix \( A \) must be equal to the number of rows in matrix \( B \) for the multiplication to be defined.

```python
#Q2
import numpy as np
import time
def mat_mult(a, b):
    result = [[0 for _ in range(len(b[0]))] for _ in range(len(a))]

    for i in range(len(a)):
        for j in range(len(b[0])):
            for k in range(len(b)):
                result[i][j] += a[i][k] * b[k][j]

    return result

# Example
mat_a = [[1, 2, 3],[4, 5, 6],]
mat_b = [[7, 8],[9, 10], [11, 12],]

result = mat_mult(mat_a, mat_b)
print("Result using list of lists:")
for row in result:
    print(row)
result_numpy = np.dot(mat_a, mat_b)
print("\nResult using NumPy:")
print(result_numpy)

#using list of list time
start_time_list_of_lists = time.time()
result_list_of_lists = mat_mult(mat_a, mat_b)
end_time_list_of_lists = time.time()
elapsed_time_list_of_lists = end_time_list_of_lists - start_time_list_of_lists

#NumPy
start_time_numpy = time.time()
result_numpy = np.dot(mat_a, mat_b)
end_time_numpy = time.time()
elapsed_time_numpy = end_time_numpy - start_time_numpy

print("\nTiming Comparison:")
print(f"List of Lists: {elapsed_time_list_of_lists} seconds")
print(f"NumPy: {elapsed_time_numpy} seconds")

#Numpy is comparatively faster as it is computed in c which makes it faster to python lists
#Numpy is collection of similar data-types densely packed in memory and is able to divide the tasks in multiple subtasks. 
```

```python
#Q9
import matplotlib.pyplot as plt
import numpy as np
def f1(x):
    return x

def f2(x):
    return x**2

def f3(x):
    return (x**3)/100

def f4(x):
    return np.sin(x)

def f5(x):
    return np.sin(x)/x

def f6(x):
    return np.log(x)
def f7(x):
    return np.exp(x)
    
x_values = np.arange(0.5, 100.5, 0.5)

plt.subplot(4, 2, 1)
plt.plot(x_values, f1(x_values), label='y=x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y=x')
plt.legend()
plt.grid(True)

plt.subplot(4, 2, 2)
plt.plot(x_values, f2(x_values), label='y=x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y=x^2')
plt.legend()
plt.grid(True)

plt.subplot(4, 2, 3)
plt.plot(x_values, f3(x_values), label='y=x^3/100')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y=x^3/100')
plt.legend()
plt.grid(True)

plt.subplot(4, 2, 4)
plt.plot(x_values, f4(x_values), label='y=sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y=sin(x)')
plt.legend()
plt.grid(True)

plt.subplot(4, 2, 5)
plt.plot(x_values, f5(x_values), label='y=sin(x)/x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y=sin(x)/x')
plt.legend()
plt.grid(True)

plt.subplot(4, 2, 6)
plt.plot(x_values, f6(x_values), label='y=log(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y=log(x)')
plt.legend()
plt.grid(True)

plt.subplot(4, 2, 7)
plt.plot(x_values, f7(x_values), label='y=e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y=e^x')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Q13
import numpy as np

def argmin(arr):
    min_index = 0
    for i in range(1, len(arr)):
        if arr[i] < arr[min_index]:
            min_index = i

    return min_index
    
arr_test = np.array([5, 3, 8, 1, 7])
result = argmin(arr_test)

result_np = np.argmin(arr_test)

print("Argmin result:", result)
print("np.argmin result:", result_np)

# Check if both results match
assert result == result_np, "Results do not match!"
```
